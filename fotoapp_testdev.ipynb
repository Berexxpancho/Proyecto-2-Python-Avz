```python
from fotoapp import *
from PIL import Image

# imagen
while True:
    ruta = input("Ingrese el nombre de la  imagen a probar: ")
    try:
        Image.open(ruta)
        break
    except:
        print("Error: No se pudo abrir la imagen. Intente nuevamente.\n")

# plataforma
plataformas_validas = ["youtube", "instagram", "twitter", "facebook"]
print("\nPlataformas disponibles:", ", ".join(plataformas_validas))

while True:
    plataforma = input("Ingrese la plataforma: ").lower()
    if plataforma in plataformas_validas:
        break
    print("Error: Plataforma inválida. Intente nuevamente.\n")

# redimensionado
print("\n=== Prueba: Redimensionado ===")
img_red = redimensionar_imagen(ruta, plataforma)
print("Imagen generada:", img_red)

# contraste
print("\n=== Prueba: Contraste ===")
equalizar_contraste(img_red)

# filtro
print("\nFiltros disponibles:")
for f in FILTROS.keys():
    print(" -", f)

while True:
    filtro = input("Ingrese el filtro a aplicar: ").upper()
    if filtro in FILTROS:
        break
    print("Error: Filtro inválido. Intente nuevamente.\n")

# prueba de filtro
print("\n=== Prueba: Filtro ===")
aplicar_filtro(img_red, filtro)

# boceto
print("\n=== Prueba: Boceto modo lápiz ===")
generar_boceto(img_red, "lapiz")
```
